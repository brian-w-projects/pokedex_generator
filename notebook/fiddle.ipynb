```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
import keras
from keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint
from keras.layers import Embedding, GRU, Dense, Dropout, LSTM, BatchNormalization
from keras.models import Model, Sequential, load_model
from keras.preprocessing.text import Tokenizer
from keras.utils import to_categorical

import nltk
from nltk import ngrams
import numpy as np
import pandas as pd
import pickle
import re
import unidecode
import warnings
nltk.download('punkt')
```

<h2>Scrape Data</h2>

```python
def scrape():
    r = requests.get('https://pokemondb.net/pokedex/all')
    soup = BeautifulSoup(r.content, 'html.parser')
    for a in set(soup.select('a.ent-name', href=True)[62:]):
        if 'pokedex' in a['href']:
            print('https://pokemondb.net'+a['href'])
            pokedex = BeautifulSoup(requests.get('https://pokemondb.net'+a['href']).content, 'html.parser')
            yield list(set([p.get_text() for p in pokedex.select('.cell-med-text')]))
```

```python
data = []
for i in scrape():
    data.extend(i)
    time.sleep(1)
```

<h2>Create Model</h2>

```python
data = pd.read_csv('data.csv')
data = list(data.iloc[:, 1])
```

```python
def tokenize(data):
    for p in data:
        yield "<START> " + " ".join(re.sub(r"[.,'()\"!-]", "", unidecode.unidecode(p)).lower().split()) + " <END>"
```

```python
tokenizer = Tokenizer(filters="")
tokenizer.fit_on_texts(i for i in tokenize(data))
```

```python
vocab_size = len(tokenizer.word_index.items()) + 1
seed_length = 2
```

```python
all_grams = []
lines = list(set([i for i in tokenize(data)]))

for line in lines:
    for gram in ngrams(line.split(" "), seed_length+1):
        all_grams.append(list(gram))
        
```

```python
sequences = np.array(tokenizer.texts_to_sequences(all_grams))
X, y = sequences[:, :-1], sequences[:, -1]
y = to_categorical(y, num_classes=vocab_size)
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 128, input_length=seed_length))
model.add(LSTM(128, return_sequences=True, dropout=0.2, recurrent_dropout=0.2))
model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
model.add(BatchNormalization())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(vocab_size, activation='softmax'))
```

```python
callbacks = [
    ReduceLROnPlateau(monitor='loss', factor=0.1, patience=10),
    EarlyStopping(monitor='loss', patience=20, restore_best_weights=True),
    ModelCheckpoint('model.h5', monitor='loss', save_best_only=True, )
]
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, batch_size=128, epochs=100, callbacks=callbacks)
```

<h2>Make Predictions</h2>

```python
def reweight_distribution(original_distribution, temperature=0.5):
    with warnings.catch_warnings():
        warnings.simplefilter('ignore')
        distribution = np.log(original_distribution) / temperature
        distribution = np.exp(distribution)
        return distribution / np.sum(distribution)
```

```python
seed = ['<start>', 'this']
temp = .5
dictionary = [""] + list(tokenizer.index_word.values())
start = np.array(tokenizer.texts_to_sequences(seed)).reshape(1, -1)
output = [seed[-1]]

for _ in range(45):
    results = reweight_distribution(model.predict(start), temperature=temp)
    word = np.random.choice(dictionary, size=1, p=results[0, :])[0]
    if word == '<end>': 
        if len(output) > 10:
            break
        else:
            continue
    output.append(word)
    start = np.append(start[0, 1:], tokenizer.texts_to_sequences([word])).reshape(1, -1)
print(" ".join(output))
```
